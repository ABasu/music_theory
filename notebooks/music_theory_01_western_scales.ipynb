```python
import music_theory as mt

piano = mt.Piano(span=('C', "B'"))
for tonic in range(12):
    tonic = mt.Note(tonic, tonic=tonic).get(fmt='western')
    scale = mt.Scale(tonic=tonic, stype='minor')
    scale_notes = [f"{m.strip("'."): >3}" for m in scale.get()]
    print(f"\n\033[1m{tonic: >3} Minor:\033[0m  {','.join(scale_notes)}\n")
    piano.draw(scale.get())
```
